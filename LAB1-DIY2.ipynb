3.)

```python
shape = (2, 2)
random_array = np.random.uniform(-1, 1, size=shape)

# Print the generated array
print(random_array)
```

#### 4.)


```python
# Creating Histogram

import matplotlib.pyplot as plt
plt.hist(random_array, bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Uniform Random Data')
plt.grid(True)

# Show the histogram
plt.show()
```

5.)

```python
import numpy as np
array1 = np.random.rand(10)
array2 = np.random.rand(10)
squared_diff = (array1 - array2) ** 2
sum_squared_diff = np.sum(squared_diff)

# Take the square root of the sum to get the Euclidean distance
euclidean_distance = np.sqrt(sum_squared_diff)

# Print the Euclidean distance
print("Euclidean Distance:", euclidean_distance)
```
